#### 구간합 (문제 종류)

##### 목적
배열에서 주어진 구간의 합을 빠르게 계산하는 알고리즘으로  
합배열을 사용하여 시간 복잡도를 줄이기위해 사용하는 알고리즘이다.

##### 이론
배열이 주어졌을 때, 그 배열을 바탕으로 합배열을 따로 만든다.  
예를 들면 아래와 같다.  
배열  =  [15, 13, 10, 7, 3, 12]  
합배열 = [15, 28, 38, 45, 48, 60]  

보시다시피 하나하나 더 해놓은 합배열을 미리 만들어 놓으면 이후에 구간합을 쉽게 구할 수 있다.  
예를 들어, 위 배열에서 마지막 요소인 12부터 7까지의 구간합을 구한다면  
합배열에서 해당하는 인덱스 값 두개만 빼주면 훨씬 쉽고 빠르게 할 수 있다.

##### 결론
매번 배열을 일일히 계산하지 말고 구간합에 도움이 되는 특수한 배열을 만들고  
그 배열을 바탕으로 보다 쉽고 빠르게 계산하는 것이 구간합 알고리즘의 핵심이라 할 수 있다.
